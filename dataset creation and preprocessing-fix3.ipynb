```python
#concatenating each file onto a master df
df_expend = pd.read_csv(txt_files[-1],sep='\t')
for file in txt_files[0:-1]:
    with open(file, "rb") as file_to_read: #finding each files encoding format for reading; not all were UTF-8
        result = chardet.detect(file_to_read.read())
    df = pd.read_csv(file,sep='\t',encoding = result['encoding'])
    if not df.empty:
        df['election_year'] = file[0:4]
        if file.count('Special_') == 1:  #standardizing the formatting for special elections
            df['election_type'] = file[13:-4]
        elif file.count('Sp_') == 1:
            df['election_type'] = file[8:-4]
        else:
            df['election_type'] = file[5:-4]
        df_expend = pd.concat([df_expend,df],ignore_index = True)
```

```python
#normalizing election type info
df_expend['election_type'] = df_expend['election_type'].replace('Election', 'General').replace('General_Election','General')
#sorting by election year, type, and expenditure. 
df_expend = df_expend.sort_values(by =['election_year','election_type','Total Amount'],ascending = [False,True,False])
```

### Governor name fix
In my above analysis, I found the governor name variables are not consistent: 
#### General
1. In 2010 and 2022, CanNameFirst is the governors last name, CanNameLast is the lt. gov's last name (Scott Carroll for Rick Scott, DeSantis Nunez for Ron Desantis)
2. In 1998, 2002, 2006, CanNameFirst and CanNameLast are both full names. CanNameFirst is the governor (Jeb Bush and Charlie Crist)
3. 2014, 2018 is Correct
4. after the Part 1 fix, 1998 and 2002 names are backwards
#### Primary
1. In 2002, CanNameLast is 'Not Yet Designated' while CanNameFirst is the full name  

The below solution is not perfect, as the 2010 and 2022 candidates will not have proper first names, but this issue will no longer be relevant when joining


```python
df_win_total[(df_win_total['RaceCode'] == 'GOV') & (df_win_total['CountyCode']=='BAY')]
```

```python
# Part 1: Election years 2010 and 2022
df_win_total.loc[:, 'Placeholder'] = df_win_total['CanNameLast']
gov_mask1 = ((df_win_total['RaceCode'] == 'GOV') & (df_win_total['election_year'].isin(['2010', '2022']))
             & (df_win_total['election_type'] != 'Primary') )
df_win_total.loc[gov_mask1, 'CanNameLast'] = df_win_total.loc[gov_mask1, 'CanNameFirst']
df_win_total.loc[gov_mask1, 'CanNameFirst'] = df_win_total.loc[gov_mask1, 'Placeholder']
df_win_total.drop(columns = 'Placeholder',inplace=True)

# Part 2: Election years 1998, 2002, 2006
gov_mask2 = ((df_win_total['RaceCode'] == 'GOV') & (df_win_total['election_year'].isin(['1998', '2002', '2006'])) 
             & (df_win_total['election_type'] != 'Primary') )
# Split `CanNameFirst` into `CanNameLast` and `CanNameFirst`
split_names = (
    df_win_total.loc[gov_mask2, 'CanNameFirst']
    .str.rsplit(' ', n = 1, expand = True)  # Splits into two based on last space
)
df_win_total.loc[gov_mask2, 'CanNameLast'] = split_names[0]  
df_win_total.loc[gov_mask2, 'CanNameFirst'] = split_names[1]    
```

```python
# Part 3: handling 2002 governor primary

gov_mask3 = (df_win_total['RaceCode']=='GOV') & (df_win_total['CanNameLast']== 'Not Yet Designated')

split_names = (
    df_win_total.loc[gov_mask3, 'CanNameFirst']
    .str.rsplit(' ', n = 1, expand = True)  # Splits into two based on last space
)
df_win_total.loc[gov_mask3, 'CanNameLast'] = split_names[1]  
df_win_total.loc[gov_mask3, 'CanNameFirst'] = split_names[0]    

#Part 4: backwards names in 1998 and 2002 
# Part 1: Election years 2010 and 2022
df_win_total.loc[:, 'Placeholder'] = df_win_total['CanNameLast']
gov_mask1 = ((df_win_total['RaceCode'] == 'GOV') & (df_win_total['election_year'].isin(['1998', '2002','2006']))
             & (df_win_total['election_type'] != 'Primary') )
df_win_total.loc[gov_mask1, 'CanNameLast'] = df_win_total.loc[gov_mask1, 'CanNameFirst']
df_win_total.loc[gov_mask1, 'CanNameFirst'] = df_win_total.loc[gov_mask1, 'Placeholder']
df_win_total.drop(columns = 'Placeholder',inplace=True)
```

```python
#intermediate df
df_elect_type = df_win_total[['election_type','election_stage']]

# if election_type variable contains runoff, or primary, stage will reflect that 
df_elect_type.loc[df_elect_type['election_type'].str.contains('Pri') == True, 'election_stage'] = 'Primary'
df_elect_type.loc[df_elect_type['election_type'].str.contains('Runoff') == True, 'election_stage'] = 'Runoff'

# if election_type contains gen
df_elect_type.loc[df_elect_type['election_type'].str.contains('Gen') == True, 'election_stage'] = 'General'
df_elect_type.loc[:,'election_type'] = df_elect_type.loc[:,'election_type'].str.replace('Gen_','')

# strict Primary or Runoffs in election_type are those stages of corresponding General elections
df_elect_type.loc[df_elect_type['election_type'].isin(['Primary', 'Runoff']), 'election_type'] = 'General'

# if election_type contains runoff or primary, that part of the variable will be dropped (splits var into two types)
df_elect_type.loc[:,'election_type'] = df_elect_type.loc[:,'election_type'].str.replace('Runoff_','').str.replace('Pri_','').str.replace('Primary_','')



# fixing all  special election generals to reflect "General" in election_stage
df_elect_type.loc[~df_elect_type['election_stage'].isin(['General','Primary','Runoff']),'election_stage'] = 'General'
                 

#lowering case to better reflect difference between election_type and election_stage
df_elect_type.loc[:,'election_stage'] = df_elect_type.loc[:,'election_stage'].str.lower()
```

### Using groupby to determine winner
Done by finding the max of each candidates summed vote counts, then dropping unnecessary columns.

```python
#total number of votes for each race

# Calculate the total votes per candidate within each group
df_win_total['TotalCanVotes'] = df_win_total.groupby(
    ['election_year', 'election_type','election_stage', 'RaceCode', 'Juris1num', 'Juris2num', 'CanNameFull']
)['CanVotes'].transform('sum')

# Find the maximum summed votes within each group (excluding the candidate name)
df_win_total['MaxVotesInRace'] = df_win_total.groupby(
    ['election_year', 'election_type', 'election_stage', 'RaceCode', 'Juris1num', 'Juris2num']
)['TotalCanVotes'].transform('max')

# Assign 1 to candidate with the highest vote in the race 
df_win_total['Won'] = (df_win_total['TotalCanVotes'] == df_win_total['MaxVotesInRace']).astype(int)

df_win_total['RaceVoterPool'] = df_win_total.groupby(
    ['election_year', 'election_type', 'election_stage', 'RaceCode', 'Juris1num', 'Juris2num']
)['CanVotes'].transform('sum')

##################### County by County Candidate Vote Count
df_win_total['TotalCanVotes_County'] = df_win_total.groupby(
    ['election_year', 'election_type','election_stage', 'CountyCode','RaceCode', 'Juris1num', 'Juris2num', 'CanNameFull']
)['CanVotes'].transform('sum')

df_win_total['RaceVoterPool_County'] = df_win_total.groupby(
    ['election_year', 'election_type', 'election_stage', 'CountyCode','RaceCode', 'Juris1num', 'Juris2num']
)['CanVotes'].transform('sum')
```

# Finalizing election winners - differentiating primaries and runoffs

```python
#finding party winners in the primaries

# finding the candidate with the largest amount of total votes by party for each election
df_win_total['MaxVotesInRaceByParty'] = df_win_total.groupby(
    ['election_year', 'election_type', 'election_stage','PartyCode','RaceCode', 'Juris1num', 'Juris2num']
    )['TotalCanVotes'].transform('max')

#finding winner for each party by comparing their total votes to the total votes recieved by the party
df_win_total['PartyWon'] = (df_win_total['TotalCanVotes'] == df_win_total['MaxVotesInRaceByParty']).astype(int)

#finding total number of voters for each party
df_win_total['RaceVoterPoolByParty'] = df_win_total.groupby(
    ['election_year', 'election_type','election_stage','PartyCode','RaceCode', 'Juris1num', 'Juris2num']
)['CanVotes'].transform('sum')
 
```
